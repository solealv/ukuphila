## KPIs

### PIB
Aumento 5% anual GNI

```
eje_economico = spark.table('transformed_data.eje_economico')
```

```
display(eje_economico)
```

```
economico = eje_economico.toPandas()
```

```
economico.sort_values(by=['country', 'year'], inplace = True)
```

```
kpi_pib = economico[['country', 'year', 'gdp_usd']].copy()
```

```
kpi_pib['usd_gdp_lag'] = kpi_pib['gdp_usd'].shift(1) 
```

```
kpi_pib['cambio_porcentaje'] = round((kpi_pib['gdp_usd']-kpi_pib['usd_gdp_lag'])/kpi_pib['usd_gdp_lag']*100,2)
```

```
kpi_pib.reset_index()
```

```
kpi_pib['Pais'] = kpi_pib['country'].map(dicc_pais)
kpi_pib['Continente'] = kpi_pib['country'].map(dicc_continente)
```

```
kpi_pib = kpi_pib[['Continente', 'Pais', 'year', 'gdp_usd', 'usd_gdp_lag', 'cambio_porcentaje']]
```

```
pib = spark.createDataFrame(kpi_pib)
```

```
pib.write.mode('overwrite').saveAsTable('analysis_data.kpi_pib')
```

### Salud
Disminuir 2% anual la mortalidad infantil

```
eje_salud_sp = spark.table('transformed_data.eje_salud')
eje_salud = eje_salud_sp.toPandas()
```

Mortality_rate_infant_female_per_1000_live_births
Mortality_rate_infant_male_per_1000_live_births

```
mortalidad_infantil = eje_salud[['country', 'year', 'Mortality_rate_infant_female_per_1000_live_births', 'Mortality_rate_infant_male_per_1000_live_births']].copy()
```

```
mortalidad_infantil.sort_values(by=['country', 'year'], inplace = True)
```

```
mortalidad_infantil['mortalidad_media'] = (mortalidad_infantil.Mortality_rate_infant_female_per_1000_live_births + mortalidad_infantil.Mortality_rate_infant_male_per_1000_live_births)/2 
```

```
mortalidad_infantil['mortalidad_media_lag'] = mortalidad_infantil['mortalidad_media'].shift(1)
```

```
mortalidad_infantil['variacion_anual'] = (mortalidad_infantil['mortalidad_media'] - mortalidad_infantil['mortalidad_media_lag'])/ mortalidad_infantil['mortalidad_media_lag']*100 # Variacion porcentual de la mortalidad infantil respecto al año anterior
```

```
mortalidad_infantil.reset_index()
```

```
mortalidad_infantil['Pais'] = mortalidad_infantil['country'].map(dicc_pais)
mortalidad_infantil['Continente'] = mortalidad_infantil['country'].map(dicc_continente)
mortalidad_infantil = mortalidad_infantil[['Continente', 'Pais','year', 'Mortality_rate_infant_female_per_1000_live_births', 'Mortality_rate_infant_male_per_1000_live_births', 'mortalidad_media', 'mortalidad_media_lag', 'variacion_anual']]
mortalidad_infantil
```

```
mortalidad = spark.createDataFrame(mortalidad_infantil)
mortalidad.write.mode('overwrite').saveAsTable('analysis_data.mortalidad_infantil')
```

###Ambiental
Aumentar un 5% anual el índice de desempeño ambiental

```
ambiental = spark.table('transformed_data.eje_ambiental_pais')
eje_ambiental = ambiental.toPandas()
```

```
epi = eje_ambiental[['country', 'EPI.new', 'EPI.old']].copy()
```

```
epi[['EPI.new', 'EPI.old']] = epi[['EPI.new', 'EPI.old']].astype(float)
```

```
epi['variacion_epi'] = ((epi['EPI.new'] - epi['EPI.old'])/epi['EPI.old']*100)/10 # Promedio de la variacion porcentual anual de los ultimos 10 anios del EPI, donde EPI.new es el valor del EPI del 2022 y EPI.old es el valor del EPI del 2012, diez anios antes.
```

```
epi['Pais'] = epi['country'].map(dicc_pais)
epi['Continente'] = epi['country'].map(dicc_continente)
epi = epi[['Continente', 'Pais', 'EPI.new', 'EPI.old', 'variacion_epi']]
epi
```

```
kpi_epi = spark.createDataFrame(epi)
kpi_epi.write.mode('overwrite').saveAsTable('analysis_data.kpi_epi')
```
